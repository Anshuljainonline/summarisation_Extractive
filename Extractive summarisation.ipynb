```python
import re
def replace_contractions(text): ## to replact constractions
    if True:
        text = text.split()
        new_text = []
        for word in text:
            if word in contractions:
                new_text.append(contractions[word])
            else:
                new_text.append(word)
        text = " ".join(new_text)
        text = re.sub(r'\'', "'", text)
        text = re.sub(r'[_"\-;%()|+&=*%,!?:#$@\[\]/]', ' ', text)
    return text

def give_string(df1,productID):  ## to get string of all the resievs combined by productID(asin)
    df1['reviewText']=df1['reviewText'].apply(replace_contractions)
    df2= df1[df1['asin']==productID]
    
    lis=[''.join(x) for x in df2.reviewText]
    str=""
    for review in lis:
        str= str + ' ' + review
    str=str.lower()
    return str
    
```

```python
productID='B0000AXRH5'
strin= give_string(df1,productID)
print(strin) ## All reviews related to the particular product ID
```

```python
short_sum=summarize(strin,ratio=0.05) # By ratio
print(short_sum)
```

```python
productID='B00002243X'
strin2= give_string(df1,productID)
short_sum2=summarize(strin2,word_count=100) #by word count
print(short_sum2)
```

```python
## LsaSummarizer
from sumy.summarizers.lsa import LsaSummarizer
parser=PlaintextParser.from_string(strin,Tokenizer('english'))

lsa_summarizer=LsaSummarizer()
lsa_summary= lsa_summarizer(parser.document,sentences_count=3)

for sentence in lsa_summary:
    print(sentence)
```

## Abstractive Summary

```python
from transformers import T5Tokenizer, T5Config, T5ForConditionalGeneration
my_model = T5ForConditionalGeneration.from_pretrained('t5-small')
tokenizer = T5Tokenizer.from_pretrained('t5-small')
```

```python
tokenizer
```

```python
text = "summarize:" + strin
text = text.strip().replace("\n","")
text
```

```python
input_ids=tokenizer.encode(text, return_tensors='pt', max_length=512)
summary_ids = my_model.generate(input_ids,num_beams=4,
                                    no_repeat_ngram_size=2,
                                    min_length=40,
                                    #max_length=150,
                                    early_stopping=True)
print(summary_ids)
```

```python
t5_summary = tokenizer.decode(summary_ids[0])
print(t5_summary)
```
